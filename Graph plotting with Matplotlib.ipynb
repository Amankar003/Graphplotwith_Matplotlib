```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,10])
y = np.array([0,300])
```

```python
x
```

```python
y
```

## 1st Graph Plotting

```python
plt.plot(x,y)
plt.show()
```

## Putting labels on the graph

```python
plt.xlabel('frequency')
plt.ylabel('count')
plt.plot(x,y)
plt.show()
```

## Changing shape of the graph

```python
plt.xlabel('frequency')
plt.ylabel('count')
plt.plot(x,y, 'o')
plt.show()
```
